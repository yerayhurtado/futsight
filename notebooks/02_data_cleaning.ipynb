# 02. Limpieza de Datos (Data Cleaning) 🧹

Este notebook se dedica a refinar el dataset fusionado (`final_dataset.csv`) para asegurar su calidad antes del Análisis Exploratorio de Datos (EDA) y la aplicación de algoritmos de Machine Learning.

Los pasos principales son:
1.  **Carga y Verificación Inicial:** Cargar el dataset y comprobar nulos/duplicados.
2.  **Conversión de Tipos:** Transformar columnas de conteo (`float`) a enteros (`Int64`).
3.  **Estandarización de Texto:** Limpiar y uniformizar columnas categóricas (`Squad`, `League`, etc.).

---

## 🛠️ Configuración e Inspección Inicial

```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/raw/final_dataset.csv')

print(f"Dataset cargado: {len(df)} filas.")

# 1. Inspección de Duplicados (Chequeo de Consistencia)
print(f"Registros duplicados: {df.duplicated().sum()}")
```

# Lista de columnas que representan CONTEOS, IDs y Premios (deben ser enteros)
int_cols = [
    'PlayerID', 'Born', 'Age', 'MP', 'Min', 'Starts', 'Subs', 'unSub', 
    'Gls', 'Ast', 'PK', 'PKatt', 'OG', 'TklW', 'Int', 'Blocks', 'Clr', 'Fls', 
    'Recov', 'Off', 'Dis', 'Won', 'Lost', 
    'Ballon d or', 'European Golden Shoe', 'League Won', 'UCL_Won', 
    'The Best FIFA Mens Player', 'UEFA Best Player', 
    'UCL_MP', 'UCL_Gls', 'market_value_in_eur' 
]

print("Realizando conversión de tipos...")
for col in int_cols:
    try:
        # 1. Rellenar posibles NaN con -1 temporalmente (solo si hubiera nulos)
        # 2. Redondear a cero y convertir a Int64
        df[col] = df[col].fillna(-1).round(0).astype('Int64')

```python
int_cols = [
    'Born', 'Age', 'MP', 'Min', 'Starts', 'Subs', 'unSub', 
    'Gls', 'Ast', 'PK', 'PKatt', 'OG', 'TklW', 'Int', 'Blocks', 'Clr', 'Fls', 
    'Recov', 'Off', 'Dis', 'Won', 'Lost', 
    'Ballon d or', 'European Golden Shoe', 'League Won', 'UCL_Won', 
    'The Best FIFA Mens Player', 'UEFA Best Player', 
    'UCL_MP', 'UCL_Gls', 'market_value_in_eur' 
]

for col in int_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int)

print("\n✅ Tipos de datos de conteo convertidos a Int64.")
print("\nVerificación de tipos después de la conversión:")
print(df[['MP', 'Gls', 'market_value_in_eur', 'Age']].dtypes)

```

## 2. Estandarización de Variables Categóricas

Estandarizamos las columnas de texto a **minúsculas** y **eliminamos espacios en blanco** para asegurar que los conteos y agrupaciones sean precisos (ej. "Arsenal " vs "arsenal").

```python
# Lista de columnas categóricas de texto
object_cols = ['Player', 'Squad', 'League', 'Nation', 'Pos', 'Season']

for col in object_cols:
    # Convertir a string, eliminar espacios en blanco y poner en minúsculas
    df[col] = df[col].astype(str).str.strip().str.lower()

# Verificación de un ejemplo
print("\n--- Conteo de Valores Únicos en Posición (Pos) ---")
print(df['Pos'].value_counts())

print("\n✅ Columnas categóricas estandarizadas.")
```

## 3. Exportación del Dataset Limpio

Guardamos el DataFrame final después de todas las correcciones de tipos y estandarización para que los siguientes notebooks (EDA y Modelado) trabajen con datos de máxima calidad.

```python
df.to_csv('../data/processed/final_dataset_cleaned.csv', index=False)
print(f"\n🎉 EXPORTACIÓN FINALIZADA.")
```
